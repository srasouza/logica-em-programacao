# Operadores Lógicos



*   And

*   Not






### Tabela Verdade

And

```python
print(16 > 6 and 2 < 5) # V and V - V
print(16 > 6 and 2 > 5) # V and F - F
print(16 < 6 and 2 > 5) # F and V - F
print(16 < 6 and 2 < 5) # F and F - F
```

Or

```python
print (5 > 3 or 1 < 2) # V or V - V
print (5 > 3 or 1 > 2) # V or F - V
print (5 < 3 or 1 < 2) # F or V - V
print (5 < 3 or 1 > 2) # F or F - F
```

Not

```python
not 16 > 6 # V - F
```

```python
not 16 < 6 # F - V
```

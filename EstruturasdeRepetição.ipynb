# Estrutura de repetição

```python
# Lista de notas de alunos
```

### For

```python
# For Range
for (contador) in range(15):
  print(contador)
```

```python
# For Range + Break
for contador in range(200):
  print(contador)
  if contador == 8:
    break
```

```python
lista = ["Teste1", "Teste2"]
```

```python
# For Simples
for vou in lista:
  print(vou)
```

```python
# For enumerate
for a, gol in enumerate (lista):
  print(a, gol)
```

### For com Operadores Compostos

```python
lista2 = [5,8,9]
```

```python
for vou in lista2 :
  vou +=2
  print(vou)
```

```python
for vou in lista2:
  vou *=3
  print(vou)
```

### While

```python
vou = 1

while vou < 9:
  vou *=5
  print(vou)
```

```python
# While com else
vou = 2
while vou < 5:
  vou *=3
else:
  print('O valor de vou é', vou)
```

```python
# Lista de notas do alunos

lista3 = [65,67,53,45,85]

# Calcular média
soma = 0
for nota in lista3:
  soma += nota
media = soma/len(lista3)

# Classificar as notas
alunos_aprovados = []
alunos_reprovados = []

for nota in lista3:
  if nota >= media:
    alunos_aprovados.append(nota)
  else:
    alunos_reprovados.append(nota)

# Imprimir os resutados
print("Notas dos alunos:", lista3)
print("Média dos alunos:", media)
print("Alunos aprovados:", alunos_aprovados)
print("Alunos reprovados:", alunos_reprovados)
```

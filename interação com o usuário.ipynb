### Type

```python
x = 10
type(x)
print('x', type(x))

nome = 'Marcela'
print('nome', type (nome))

num = 150
num_convert = int(num) # Converter variável
print('num', type(num))
print('num_conver', type(num_convert))

num_float = float(num)
print('num_float', num_float)
```

### Input


*   Armazenar a informação digitada por um usuário em uma variável.
*   O dado armazenado sempre será um texto.

```python
nome = input('Digite o nome do usuário ')
print(nome)
```

```python
valor = input ('Digite o valor da nota ')
print(nome, 'sua nota é', valor)
```

```python
valor = input('Digite o valor do produto: ')
print(valor)
```

```python
desconto = valor - 9
```

```python
#Forma mais longa
valor_convert = int(valor)
desconto = valor_convert - 9
print(desconto)
```

```python
#Forma mais curta
desconto = int(valor) - 9
print(desconto)
```
